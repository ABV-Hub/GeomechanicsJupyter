1) Welbore problem

```python
# Import the usual suspects
import numpy as np
import pandas as pd
import seaborn as sns
import os
import glob
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedFormatter
%matplotlib inline


# 2D problem of circular cavity subject to far field effective stresses

# Working in metric units today

# Far field effective stresses
sigxx = 12 #MPa 
sigyy = 3 #MPa
# Cavity diameter
cavityDiam = 0.2 #m
# Material properties
E = 10 #GPa
v = 0.20 
# Unconfined compressive strength
ucs = 30 #MPa
# Tensile strength
ts = 2 #MPa 

# Radius of cavity
a = cavityDiam/2

# Please see FreeFEM++ code for generation of the results. Here I am simply
# pulling those results from a file to plot them later.

#2x2 m domain
XLine = pd.read_csv(r'WellboreProblemResults\output_X.dat', sep=',', header=None)
XLine.rename(columns={0:'x',1:'y',2:'sig_xx',3:'sig_yy',4:'sig_xy'},inplace=True)
YLine = pd.read_csv(r'WellboreProblemResults\output_Y.dat', sep=',', header=None)
YLine.rename(columns={0:'x',1:'y',2:'sig_xx',3:'sig_yy',4:'sig_xy'},inplace=True)
XLine = XLine[XLine['x']<1]
YLine = YLine[YLine['y']<1]

#0.5x0.5 m domain
XLine5 = pd.read_csv(r'WellboreProblemResults\output_X_2.dat', sep=',', header=None)
XLine5.rename(columns={0:'x',1:'y',2:'sig_xx',3:'sig_yy',4:'sig_xy'},inplace=True)
YLine5 = pd.read_csv(r'WellboreProblemResults\output_Y_2.dat', sep=',', header=None)
YLine5.rename(columns={0:'x',1:'y',2:'sig_xx',3:'sig_yy',4:'sig_xy'},inplace=True)
XLine5 = XLine5[XLine5['x']<.25]
YLine5 = YLine5[YLine5['y']<.25]
```

    a. Use Kirch equations to compute sigrr, sigthth,sigrth. How far does the presence of the wellbore influence stresses?

```python
def Kirsch(sig1,sig2,a,r,th,Pw=0):
    """Function defining Kirsch equations """
    A = (sig1+sig2-2*Pw)/2
    B = (sig1-sig2)/2
    sigRR= A*(1- (a**2)/(r**2)) + B*(1-(2*a/r)**2 
                                     + 3*(a**4)/(r**4))*np.cos(2*th) \
                                    +Pw*(a/r)**2
    sigTT= A*(1+ (a**2)/(r**2)) - B*(1+ 3*(a**4)/(r**4))*np.cos(2*th)\
                                    -Pw*(a/r)**2
    sigRT= -B*(1+2*(a/r)**2 - 3*(a**4)/(r**4))*np.sin(2*th)
    return(sigRR,sigTT,sigRT)

#Create cartesian mesh
s = 1
meshSize = 20
y=np.linspace(-s,s,meshSize)
x=np.linspace(-s,s,meshSize)
xG,yG = np.meshgrid(x,y)
xG = xG.reshape(meshSize**2)
yG = yG.reshape(meshSize**2)

#Convert cartesian mesh to polar mesh
r_toss = np.sqrt(xG**2 + yG**2)

#Toss any points inside the wellbore
outside = r_toss > a
xG=xG[outside]
yG=yG[outside]
r = np.sqrt(xG**2 + yG**2)
th = np.arctan2(yG,xG)


#Calculate stresses along mesh
(sigRR,sigTT,sigRT) = Kirsch(sigxx,sigyy,a,r,th)

def stressHeatMap(xG,yG,sig, name,axes):
    df = {"x [m]":xG,"y [m]":yG,"z":sig}
    df = pd.DataFrame(df)
    d = df.pivot("x [m]","y [m]","z")
    ax= sns.heatmap(d, xticklabels=d.columns.values.round(2),
                    yticklabels=d.columns.values.round(2),
                    ax = axes)
    ax.set_title(name)
    return ax
```

```python
f, axes = plt.subplots(1, 3, figsize=(15, 4))
ax1 = stressHeatMap(xG,yG,sigRR, '$\sigma_{r,r}$ [MPa]', axes[0])
ax2 = stressHeatMap(xG,yG,sigRT, '$\sigma_{r,\Theta}$ [MPa]', axes[1])
ax3 = stressHeatMap(xG,yG,sigTT, '$\sigma_{\Theta \Theta}$ [MPa]', axes[2])
```

b. Compute and plot stresses in a line. I will combine this part with the FreeFEM comparison from part d to save on space.

```python
x = np.linspace(0.1,1,20)
y = np.zeros(np.shape(x))
rL = np.sqrt(x**2 + y**2)
thL = y
(sigRRL,sigTTL,sigRTL) = Kirsch(sigxx,sigyy,a,rL,thL)

xF=XLine['x']
yF=XLine['y']
sigRRF=-XLine['sig_xx']/1e6
sigRTF=-XLine['sig_xy']/1e6
sigTTF=-XLine['sig_yy']/1e6



#Plot results
f, axes = plt.subplots(1, 3, figsize=(15, 4))
ax1 = axes[0].plot(x,sigRRL, '-o', label = 'Analytical')
ax1 = axes[0].plot(xF,sigRRF, '-o',label = 'FEM')
axes[0].set_title('$\sigma_{r,r}$ [MPa]')
axes[0].grid()
axes[0].legend()

ax2 = axes[1].plot(x,sigRTL, '-o', label = 'Analytical')
ax2 = axes[1].plot(xF,sigRTF, '-o', label = 'FEM')
axes[1].set_title('$\sigma_{r,\Theta}$ [MPa]')
axes[1].grid()
axes[1].legend()

ax3 = axes[2].plot(x,sigTTL, '-o', label = 'Analytical')
ax3 = axes[2].plot(xF,sigTTF, '-o', label = 'FEM')
axes[2].set_title('$\sigma_{\Theta,\Theta}$ [MPa]')
axes[2].grid()
axes[2].legend()

t = f.suptitle('Kirsch Analytic Solution',fontsize=16)
t.set_position([.5, 1.05])
```

```python
# Repeating the process for a domain of 0.5 x 0.5 m
# Let us compare Kirsch's results to our freefem++ results
xF2=XLine5['x']
yF2=XLine5['y']
sigRRF2=-XLine5['sig_xx']/1e6
sigRTF2=-XLine5['sig_xy']/1e6
sigTTF2=-XLine5['sig_yy']/1e6


#Plot results, don't forget to set bounds so colors are comparable across our python and freefem plots
f, axes = plt.subplots(1, 3, figsize=(15, 4))
ax1 = axes[0].plot(xF,sigRRF, '-o',label = 'FEM')
ax1 = axes[0].plot(xF2,sigRRF2, '-o',label = 'FEM2')
axes[0].set_title('$\sigma_{r,r}$ [MPa]')
axes[0].grid()
axes[0].legend()

ax2 = axes[1].plot(xF,sigRTF, '-o', label = 'FEM')
ax2 = axes[1].plot(xF2,sigRTF2, '-o', label = 'FEM2')
axes[1].set_title('$\sigma_{r,\Theta}$ [MPa]')
axes[1].grid()
axes[1].legend()

ax3 = axes[2].plot(xF,sigTTF, '-o', label = 'FEM')
ax3 = axes[2].plot(xF2,sigTTF2, '-o', label = 'FEM2')
axes[2].set_title('$\sigma_{\Theta,\Theta}$ [MPa]')
axes[2].grid()
axes[2].legend()
t = f.suptitle('Freefem++ Domain Comparison',fontsize=16)
t.set_position([.5, 1.05])
```

## It would appear that the magnitude and trend of our analytic model tracks the FreeFEM solution nicely.

## Also, shrinking the domain size with the same mesh resulted in a poor quality simulation. This was most likely because in the second simulation, by shrinking the domain size we also shrank the distance between the wellbore and the boundaries. This means that the farfield stresses we applied at the boundaries had a much more significant effect than they should have.

```python
y = np.linspace(0.1,1,20)
x = np.zeros(np.shape(y))
rL = np.sqrt(x**2 + y**2)
thL = np.ones(np.shape(y))*np.pi/2
(sigRRL,sigTTL,sigRTL) = Kirsch(sigxx,sigyy,a,rL,thL)


xF=YLine['x']
yF=YLine['y']
sigRRF=-YLine['sig_yy']/1e6
sigRTF=-YLine['sig_xy']/1e6
sigTTF=-YLine['sig_xx']/1e6



#Plot results, don't forget to set bounds so colors are comparable across our python and freefem plots
f, axes = plt.subplots(1, 3, figsize=(15, 4))
ax1 = axes[0].plot(y,sigRRL, '-o', label = 'Analytical')
ax1 = axes[0].plot(yF,sigRRF, '-o',label = 'FEM')
axes[0].set_ylabel('$\sigma_{r,r}$ [MPa]')
axes[0].grid()
axes[0].legend()

ax2 = axes[1].plot(y,sigRTL, '-o', label = 'Analytical')
ax2 = axes[1].plot(yF,sigRTF, '-o', label = 'FEM')
axes[1].set_ylabel('$\sigma_{r,\Theta}$ [MPa]')
axes[1].grid()
axes[1].legend()

ax3 = axes[2].plot(y,sigTTL, '-o', label = 'Analytical')
ax3 = axes[2].plot(yF,sigTTF, '-o', label = 'FEM')
axes[2].set_ylabel('$\sigma_{\Theta,\Theta}$ [MPa]')
axes[2].grid()
axes[2].legend()
plt.tight_layout()
t = f.suptitle('Kirsch Analytic Solution',fontsize=16)
t.set_position([.5, 1.05])
```

c. Use Kirsch equations to compute sig_rr and sig_tt for r=0.1. Is there any section of rock in shear or tensile failure? Where?

```python
th = np.linspace(0,2*np.pi,20)
th = th[0:-1]
r = np.ones(np.shape(th))*0.1

(sigRRc,sigTTc,sigRTc) = Kirsch(sigxx,sigyy,a,r,th)
x=r*np.cos(th)
y=r*np.sin(th)


def circleStressHeatMap(xG,yG,sig, name,axes):
    df = {"x [m]":xG.round(2),"y [m]":yG.round(2),"z":sig}
    df = pd.DataFrame(df)
    d = df.pivot("x [m]","y [m]","z")
    ax= sns.heatmap(d, xticklabels=yG.round(2),ax=axes)
    ax.set_title(name)
    return ax

f, axes = plt.subplots(1, 2, figsize=(15, 4))
ax1 = circleStressHeatMap(x,y,sigRRc, '$\sigma_{r,r}$ [MPa]', axes[0])
ax3 = circleStressHeatMap(x,y,sigTTc, '$\sigma_{\Theta \Theta}$ [MPa]', axes[1])
```

```python
# Tensile failure should be expected wherever the hoop stress, sig_theta,theta drops is less than -TensileStrength.
# Shear failure should be expected wherever te hoop stress is above the UCS

tensile = sigTTc<=-ts
shear = sigTTc>=ucs

fig= plt.figure(figsize=(5,5))
ax=fig.add_subplot(111)
ax.plot(x,y,'k.')
ax.plot(x[tensile],y[tensile],'ro',MarkerSize=9,label='Tensile Failure Likely')
ax.plot(x[shear],y[shear],'go',MarkerSize=9,label='Shear Failure Likely')
ax.legend()
ax.set_title("Wellbore Wall")
ax.grid()
plt.axis("square")
```

```python
# The next problem is to analyze a fracture in a formation, part a was done in freefem++. Here we review those results
# and perform some comparison with the analytical solutions
# Note that I chose to take advantage of the symmetry of this problem and
# only simulate half of the domain.

#Setup subplots
fig, ax = plt.subplots()


#Part b, Plot sig_xx perpendicular to fracture. How far does 
#influence of the fracture go?
fracData = pd.read_csv(r'FractureProblemResults\partb.dat', sep=',', header=None)
fracData.rename(columns={0:'x',1:'y',2:'sig_xx',3:'sig_yy',4:'sig_xy'},inplace=True)

xb = fracData['x']
sigTTb = -fracData['sig_xx']/1e6

ax.plot(xb,sigTTb,'-o',label="FEM")
ax.set_xlabel("x coordinate [m]")
ax.set_ylabel("Sigma XX [MPa]")
ax.set_title("Stress perpendicular to fracture")
ax.grid()
```

## The significant influence of the fracture appears to die out after 30 meters.

Part c, Plot x-displacements at the face of the fracture. Compare with analytical equation

```python
# Analytical solution
c=10
w0=0.1/2
yA = np.linspace(50.1,100,20)
xA = np.zeros(np.shape(yA))

p0 = 10
E= 30*(1e3)
v= 0.2
Ep = E/(1-v**2)

#Grab freefem results
fracData = pd.read_csv(r'FractureProblemResults\partc.dat', sep=',', header=None)
fracData.rename(columns={0:'x',1:'y',2:'u1',3:'u2'},inplace=True)
xc = fracData['x']
yc = fracData['y']
xDisplacement = fracData['u1']

#Calculate stress on face of fracture using analytical solution
y = np.linspace(-c,c,20)
u_xA = 2*(p0/Ep)*np.sqrt(c**2 - (y)**2)
y+=c

#Plot results
fig, ax = plt.subplots()
ax.plot(yc-(50-c),xDisplacement,label="FEM")
ax.plot(y,u_xA,'o',label="Analytical")
ax.legend()
ax.grid()
ax.set_xlabel("Point along fracture surface [m]")
ax.set_ylabel("X - Displacement [m]")
plt.tight_layout()
```
