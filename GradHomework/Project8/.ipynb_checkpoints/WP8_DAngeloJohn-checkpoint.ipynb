```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
reversalList = []

def findReverse(df):
    sXX=df['sig_xx'].tolist()
    sYY=df['sig_yy'].tolist()
    reversal=0
    for i in range(len(sXX)):
        if sXX[i]>=sYY[i]:
            reversal = i
            break
    return reversal
```

1) Stress shadows and interference

```python
#import Freefem++ solution

v = 0.20 #poisson's ratio

#Set background stress values
Sxx=-6300*6894.76 #Shmin
Syy=-6400*6894.76 #SHmax

df = pd.read_csv("output4.dat")
pd.options.mode.chained_assignment = None

#Calculate delta sigYY from  FreeFem++ solution for 
#delta sigXX and delta sigZZ 
df['delta_sig_yy'] = v*(df['sig_xx'] + df['sig_zz'])
#Add on background stresses
df['sig_yy'] = (df['delta_sig_yy']+Syy )*1e-6
df['sig_xx'] = (df['sig_xx'] +Sxx )*1e-6

reversal= findReverse(df)

#Plot results
fig,ax=plt.subplots()
df.plot(x='x',y='sig_xx',ax=ax)
df.plot(x='x',y='sig_yy',ax=ax)
# df.plot(x='x',y='sig_zz',ax=ax)
ax.grid()
ax.set_ylabel('Stress [MPa]')
ax.set_xlabel('x [m]')
ax.plot(df['x'].iloc[reversal],df['sig_xx'].iloc[reversal],'ro')
ax.legend()
reversalList.append([4,df['x'].iloc[reversal]])
```

2) Coupled hydraulic fracturing modeling

```python
#Given parameters
hf = 170 #hydraulic fracture height, ft
E_ps = 8.9 #plane-strain modulus, Mpsi
inj_rate = 50 #injection rate, bbl/min <-- Held constant for 1 hour
```

```python
#Convert parameters to SI
hf = 170*0.3048 #hydraulic fracture height, m
E_ps = 8.9*6895*(10**6) #plane-strain modulus, Pa
inj_rate = 50*(6.309e-5)  #injection rate, m3/s <-- Held constant for 1 hour
```

a. Compute the expected fracture half-length xf, fracture width at wellbore w_w0, and net pressure pn as a function of time with the PKN model (no leaf off) assuming fracturing fluid has constant viscosity, 2cP.

```python
mu = 2 #viscosity of fracture fluid in cP
mu = 2*0.001 #viscosity of fracture fluid in Pa*s
t_span = 60*60 #told that injection goes on for 1 hour, so this is 1 hr in seconds

#Fracture half length in meters
xf = 0.524*(((inj_rate**3)*E_ps/mu/(hf**4))**(1/5))*(t_span**(4/5))

#Maximum width at the wellbore in meters
w_w0 = 3.04*(((inj_rate**2)*mu/E_ps/hf)**(1/5))*(t_span**(1/5))

#Net Pressure at the wellbore in Pa
p_n = 1.52*((((E_ps**4)*(inj_rate**2)*mu/(hf**6))**(1/5))*(t_span**(1/5)))

print("Fracture half length:")
print(xf," [m]")
print("Maximum width at the wellbore:")
print(w_w0, " [m]")
print("Net Pressure at the wellbore:")
print(p_n/(10**6), " [MPa]")
```

b. Compute the expected fracture half-length xf, fracture width at the wellbore w_w0, and
net pressure pn as a function of time with the PKN model (no leak-off). Now
the fracturing fluid has a viscosity 2 cP with no proppant (initial 10 min), and
increases in steps of 10 min with 2 cP in each step (due to increasing
proppant concentration).

```python
#Generate requested interval of viscosity values
mu_interval = 2*0.001 
mu_t = mu_interval
mu_data = []
for t in range(3600):
    if t < 10*60:
        mu_data.append(mu_t)
    else:
        if t%600 == 0:
            mu_t += mu_interval            
            
        mu_data.append(mu_t)
mu_data=np.array(mu_data)
t_span = np.arange(len(mu_data))
t_interval = t_span%10
#Fracture half length in meters
xf = 0.524*(((inj_rate**3)*E_ps/mu_data/(hf**4))**(1/5))*(t_interval**(4/5))
xf=np.cumsum(xf)
#Maximum width at the wellbore in meters
w_w0 = 3.04*(((inj_rate**2)*mu_data/E_ps/hf)**(1/5))*(t_interval**(1/5))
w_w0=np.cumsum(w_w0)
#Net Pressure at the wellbore in Pa
p_n = 1.52*((((E_ps**4)*(inj_rate**2)*mu_data/(hf**6))**(1/5))*(t_interval**(1/5)))
p_n=np.cumsum(p_n)
fig,ax=plt.subplots()
ax.plot(t_span/60,mu_data)
ax.set_xlabel("Time [min]")
ax.set_ylabel("Viscosity [Pa*s]")
ax.grid()
```

```python
fig,axs=plt.subplots(3,1,figsize=(8,10))
axs[0].plot(t_span,xf)
axs[0].set_xlabel('Time [s]')
axs[0].set_ylabel('Fracture Half Length [m]')
axs[0].grid()
axs[1].plot(t_span,w_w0*1000)
axs[1].set_xlabel('Time [s]')
axs[1].set_ylabel('Maximum width at wellbore [mm]')
axs[1].grid()

axs[2].plot(t_span,p_n/(10**6))
axs[2].set_xlabel('Time [s]')
axs[2].set_ylabel('Net Pressure at Wellbore [MPa]')
axs[2].grid()

plt.tight_layout()
```

c. What should you do to your solution in order to consider leak-off? Justify and
explain briefly the algorithm to calculate xf, ww,0, and pn.

In the derivation of the PKN model, we assumed that leak off was negligible. If we want to consider leak-off, we would have to rederive the PKN model using the conservation mass, such that the volume of the fracture is influenced by not only the constant injection rate and change in fracture size, but also by leak-off (volume rate of fluid loss per unit length).  These three contributing factors should sum to zero via conservtion of mass.
There are different methods for quantifying the leak-off rate and we have not convered a particular method in class, so I will not go into this in detail. We simply assume that we now have an expression for the volume change as a function of injection rate and leak-off rate. This can be plugged into our other equations as we did in the notes. However, we will not get a nice analytical solution. We will get a differential equation that must be solved numerically. Or we could make assumptions about whether the system is storage or leak-off dominated to simplify the problem.
