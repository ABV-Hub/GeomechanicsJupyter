# Problem 1

Compute the vertical stress gradient resulting from a carbonate rock made of 70% dolomite and 30% calcite, porosity 10% and filled with brine ~1,060 kg/m3. Provide answer in psi/ft, MPa/km, and ppg.

Solving for density of carbonate rock matrix
$$\rho_{matrix}=0.7 \times 2.87 + 0.3 \times 2.71=2.822 gm/cc$$

Solving for bulk mass density of carbonate rock
$$\rho_{bulk}=0.9 \times 2.822 + 0.1 \times 1.06=2.646 gm/cc$$

Density of water is $\rho_{water}=1 gm/cc=8.33 ppg$

This corresponds to pressure gradients of $0.433 psi/ft=9.792 MPa/km$

Computing the vertical stress gradient

$$\frac{dS_{v}}{dh}=\frac{2.646 gm/cc}{1 gm/cc} \times 0.433 psi/ft = 1.146 psi/ft$$
<br>
$$\frac{dS_{v}}{dh}=\frac{2.646 gm/cc}{1 gm/cc} \times 8.33 ppg = 22.04 ppg$$
<br>
$$\frac{dS_{v}}{dh}=\frac{2.646 gm/cc}{1 gm/cc} \times 9.792 MPa/km = 25.90 MPa/km$$

# Problem 2

Calculate (without the help of a computer) the total vertical stress in an offshore location at 10,000 ft of total depth (from surface) for which: water depth is 1,000m, bulk mass density of rock at the seabed is 1,800 kg/m3 increasing linearly until a depth of 500 m below sea-floor to 2,350 kg/m3 and relatively constant after it. Why would rock bulk mass density increase with depth?

From 0 to 1000m, the vertical stress is
$$1000 kg/m^3 \times 9.81 m/s^2 \times 1000m = 9.81 MPa$$

From 1000 to 1500m, the vertical stress is
$$\frac{1800 kg/m^3 + 2350 kg/m^3}{2}  \times 9.81 m/s^2 \times 500m = 10.18 MPa$$

From 1500 to 3048m (10000ft), the vertical stress is
$$2350 kg/m^3 \times 9.81 m/s^2 \times 1548m = 36.378 MPa$$

The total vertical stress at 10000 ft is thus
$$9.81 MPa + 10.18 MPa + 36.378 MPa= 56.36 MPa = 8174 psi$$

Go to https://github.com/dnicolasespinoza/GeomechanicsJupyter and download the files ‘1_14-1_Composite.las’ and ‘1_14-1 deviation_mod.dev’. The first one is a well logging file (.LAS). You will find here measured depth (DEPTH - Track 1) and bulk mass density (RHOB - Track 8). Track 3 also shows bulk density correction (DRHO). Add RHOB to DRHO to obtain the corrected bulk mass density. The second file has the deviation survey of the well. Use this file to calculate true vertical depth subsea (TVDSS) as a function of measured depth (MD) in the well logging file. You may assume an average bulk mass density of 2 g/cc between the seafloor and the beginning of the density data. Summations with discrete data sets can be easily done through a ‘for’ loop or with a spreadsheet.

Example code of using Python to import and graph data shown below or use excel.

```python
# Written by Nathan Hsu

import matplotlib
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax

md = []
tvdss = []

with open ('1_14-1_deviation_mod.dev') as temp:
    for line in temp:
        row = line.split()
        md.append(float(row[0]))
        tvdss.append(float(row[1]))
        
#now i need to use the function i found to get TVDSS from the composite.lan file

depth = []
rhob = []
drho = []

with open ('1_14-1_Composite.las') as temp: #reading the las file
    for line in temp:
        row = line.split() #splitting into rows, note: edited the original file to delete the headings
        depth.append(float(row[0]))
        if float(row[2]) == -999.2500:
            drho.append(0.0)
        else: drho.append(float(row[2]))
        if float(row[7]) == -999.2500:
            rhob.append(2.0)
        else: rhob.append(float(row[7]))
    correctdensity = [x + y for x, y in zip(drho, rhob)]
    cdens = [x*1000 for x in correctdensity] # change g/cc to kg/m3

md = np.array(md, dtype=np.float64)
tvdss = np.array(tvdss, dtype=np.float64)
TVDSS = []

for MD in depth:   
    TVDSS.append(np.interp(MD,md,tvdss)) #now we have the true vert depth for the LAS file

vertstress = []

for i in range(len(TVDSS)-1):
    sv = (cdens[i] + cdens[i+1])/2*9.81/10**6*(TVDSS[i+1] - TVDSS[i])
    vertstress.append(sv)
    
Sv = np.cumsum(vertstress) # now we have the vertical stress for all depths
Sv = np.insert(Sv,0,0)

plt.figure(0)
plt.plot(Sv,TVDSS, 'b-', label = "S_v")
matplotlib.pyplot.grid(b=None, which='major', axis='both')
plt.title('Vertical Stress vs. Depth')
plt.xlabel('Pore pressure and stress [MPa]')
plt.ylabel('Depth [m]')
plt.gca().invert_yaxis()

plt.figure(1)
plt.plot(cdens,TVDSS, 'r-', label = "p_bulk")
matplotlib.pyplot.grid(b=None, which='major', axis='both')
plt.title('Bulk Mass Density vs. Depth')
plt.xlabel('Bulk Mass Density [kg/m3]')
plt.ylabel('Depth [m]')
plt.gca().invert_yaxis()
```
