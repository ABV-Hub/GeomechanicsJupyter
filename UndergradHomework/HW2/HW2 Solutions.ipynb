# Problem 2

Using the equations of linear elasticity show that by applying an isotropic stress $\sigma_{iso}$ (no shear) the volumetric strain is equal to $\sigma_{vol}=\frac{3(1-2v)}{E}\sigma_{iso}$. Write out the resulting strain and stress tensors.

For this problem we refer to the following coordinate system
![](files/HW_2_Problem_2_and_3_Axis.jpg)

Constitutive equation of linear elasticity is 

$$\underline{\underline{\sigma}}=\underline{\underline{\underline{\underline{C}}}}\underline{\underline{\epsilon}}$$

Or in Voigt notation

$$\underline{\sigma}=\underline{\underline{C}}\underline{\epsilon}$$

Constitutive equation can be written 

$$\underline{\underline{\epsilon}}=\underline{\underline{\underline{\underline{D}}}}\underline{\underline{\sigma}}$$

\begin{equation*}
\begin{vmatrix}
\epsilon_{11} \\
\epsilon_{22} \\
\epsilon_{33} \\
2\epsilon_{12} \\
2\epsilon_{13} \\
2\epsilon_{23} \\
\end{vmatrix}=
\begin{vmatrix}
1/E & -v/E & -v/E & 0 & 0 & 0 \\
-v/E & 1/E & -v/E & 0 & 0 & 0 \\
-v/E & -v/E & 1/E & 0 & 0 & 0 \\
0 & 0 & 0 & 2(1+v)/E & 0 & 0 \\
0 & 0 & 0 & 0 & 2(1+v)/E & 0 \\
0 & 0 & 0 & 0 & 0 & 2(1+v)/E \\
\end{vmatrix}=
\begin{vmatrix}
\sigma_{11} \\
\sigma_{22} \\
\sigma_{33} \\
\sigma_{12} \\
\sigma_{13} \\
\sigma_{23} \\
\end{vmatrix}
\end{equation*}

Isotropic stress means $\sigma_{12}$, $\sigma_{13}$, and $\sigma_{23}=0$. So we have 

\begin{equation*}
\begin{vmatrix}
\epsilon_{11} \\
\epsilon_{22} \\
\epsilon_{33} \\
2\epsilon_{12} \\
2\epsilon_{13} \\
2\epsilon_{23} \\
\end{vmatrix}=
\begin{vmatrix}
1/E & -v/E & -v/E & 0 & 0 & 0 \\
-v/E & 1/E & -v/E & 0 & 0 & 0 \\
-v/E & -v/E & 1/E & 0 & 0 & 0 \\
0 & 0 & 0 & 2(1+v)/E & 0 & 0 \\
0 & 0 & 0 & 0 & 2(1+v)/E & 0 \\
0 & 0 & 0 & 0 & 0 & 2(1+v)/E \\
\end{vmatrix}=
\begin{vmatrix}
\sigma_{iso} \\
\sigma_{iso} \\
\sigma_{iso} \\
0 \\
0 \\
0 \\
\end{vmatrix}
\end{equation*}

Matrix multiplication gives

$$\epsilon_{11}=\epsilon_{22}=\epsilon_{33}=\frac{(1-2v)}{E}\sigma_{iso}$$

Volumetric strain is

$$\epsilon_{vol}=\epsilon_{11}+\epsilon_{22}+\epsilon_{33}=\frac{3(1-2v)}{E}\sigma_{iso}$$

In summary,

\begin{equation*}
\underline{\underline{\sigma}}=
\begin{vmatrix}
\sigma_{11} & \sigma_{12} & \sigma_{13} \\
\sigma_{21} & \sigma_{22} & \sigma_{23} \\
\sigma_{31} & \sigma_{32} & \sigma_{33} \\
\end{vmatrix}=
\begin{vmatrix}
\sigma_{iso} & 0 & 0 \\
0 & \sigma_{iso} & 0 \\
0 & 0 & \sigma_{iso} \\
\end{vmatrix}
\end{equation*}

\begin{equation*}
\underline{\underline{\epsilon}}=
\begin{vmatrix}
\epsilon_{11} & \epsilon_{12} & \epsilon_{13} \\
\epsilon_{21} & \epsilon_{22} & \epsilon_{23} \\
\epsilon_{31} & \epsilon_{32} & \epsilon_{33} \\
\end{vmatrix}=
\begin{vmatrix}
\frac{(1-2v)}{E}\sigma_{iso} & 0 & 0 \\
0 & \frac{(1-2v)}{E}\sigma_{iso} & 0 \\
0 & 0 & \frac{(1-2v)}{E}\sigma_{iso} \\
\end{vmatrix}
\end{equation*}

# Problem 3

Using the equations of linear elasticity show that by applying stress in one direction (say 1) and not letting the solid expand in the other two, you can recover the following expression $\sigma_{11}=\frac{E(1-v)}{(1+v)(1-2v)}\epsilon_{11}$. The proportionality coefficient is called “M” the constrained modulus. Is it lower or higher than E? What is the physical explanation? Write out the resulting strain and stress tensors.

For this problem we refer to the following coordinate system
![](files/HW_2_Problem_2_and_3_Axis.jpg)

Constitutive equation states that $$\underline{\underline{\sigma}}=\underline{\underline{\underline{\underline{C}}}}\underline{\underline{\epsilon}}$$

\begin{equation*}
\begin{vmatrix}
\sigma_{11} \\
\sigma_{22} \\
\sigma_{33} \\
\sigma_{12} \\
\sigma_{13} \\
\sigma_{23} \\
\end{vmatrix}=\frac{E}{(1+v)(1-2v)}
\begin{vmatrix}
1-v & v & v & 0 & 0 & 0 \\
v & 1-v & v & 0 & 0 & 0 \\
v & v & 1-v & 0 & 0 & 0 \\
0 & 0 & 0 & (1-2v)/2 & 0 & 0 \\
0 & 0 & 0 & 0 & (1-2v)/2 & 0 \\
0 & 0 & 0 & 0 & 0 & (1-2v)/2 \\
\end{vmatrix}=
\begin{vmatrix}
\epsilon_{11} \\
\epsilon_{22} \\
\epsilon_{33} \\
2\epsilon_{12} \\
2\epsilon_{13} \\
2\epsilon_{23} \\
\end{vmatrix}
\end{equation*}

Applying stress in one direction (say 1) and not letting the solid expand in the other two implies $\epsilon_{22}=\epsilon_{33}=0$

So we have

\begin{equation*}
\begin{vmatrix}
\sigma_{11} \\
\sigma_{22} \\
\sigma_{33} \\
\sigma_{12} \\
\sigma_{13} \\
\sigma_{23} \\
\end{vmatrix}=\frac{E}{(1+v)(1-2v)}
\begin{vmatrix}
1-v & v & v & 0 & 0 & 0 \\
v & 1-v & v & 0 & 0 & 0 \\
v & v & 1-v & 0 & 0 & 0 \\
0 & 0 & 0 & (1-2v)/2 & 0 & 0 \\
0 & 0 & 0 & 0 & (1-2v)/2 & 0 \\
0 & 0 & 0 & 0 & 0 & (1-2v)/2 \\
\end{vmatrix}=
\begin{vmatrix}
\epsilon_{11} \\
0 \\
0 \\
0 \\
0 \\
0 \\
\end{vmatrix}
\end{equation*}

Matrix multiplication along the first row gives

$$\sigma_{11}=\frac{E(1-v)}{(1+v)(1-2v)}\epsilon_{11}$$
<br>
$$\sigma_{11}=M\epsilon_{11}$$

If we assume $v=0.2$, then $M=1.11E$. So M is greater than E since the solid shows a greater uniaxial stiffness when it is laterally constrained.

Rest of matrix multiplication (2nd and 3rd rows) gives

$$\sigma_{22}=\sigma_{33}=\frac{Ev}{(1+v)(1-2v)}\epsilon_{11}$$

Therefore, the relationship between $\sigma_{11}$ and the tectonic/horizontal stresses ($\sigma_{22}$, $\sigma_{33}$) is

$$\sigma_{22}=\sigma_{33}=\frac{v}{(1-v)}\sigma_{11}$$

Since no shear exists, we have zero shear strains ($\epsilon_{12}=\epsilon_{13}=\epsilon_{23}=0$) and zero shear stresses

$$\sigma_{12}=\frac{E}{(1+v))}\epsilon_{12}=0$$
<br>
$$\sigma_{13}=\frac{E}{(1+v))}\epsilon_{13}=0$$
<br>
$$\sigma_{23}=\frac{E}{(1+v))}\epsilon_{23}=0$$

In summary,

\begin{equation*}
\underline{\underline{\sigma}}=
\begin{vmatrix}
\sigma_{11} & \sigma_{12} & \sigma_{13} \\
\sigma_{21} & \sigma_{22} & \sigma_{23} \\
\sigma_{31} & \sigma_{32} & \sigma_{33} \\
\end{vmatrix}=
\begin{vmatrix}
\sigma_{11} & 0 & 0 \\
0 & \frac{v}{(1-v)}\sigma_{11} & 0 \\
0 & 0 & \frac{v}{(1-v)}\sigma_{11} \\
\end{vmatrix}
\end{equation*}

\begin{equation*}
\underline{\underline{\epsilon}}=
\begin{vmatrix}
\epsilon_{11} & \epsilon_{12} & \epsilon_{13} \\
\epsilon_{21} & \epsilon_{22} & \epsilon_{23} \\
\epsilon_{31} & \epsilon_{32} & \epsilon_{33} \\
\end{vmatrix}=
\begin{vmatrix}
\frac{1}{M}\sigma_{11} & 0 & 0 \\
0 & 0 & 0 \\
0 & 0 & 0 \\
\end{vmatrix}
\end{equation*}

# Problem 4
<br>
The top of the Barnett shale is located at about 7950 ft TVD. At this depth (assume $v=0.22$):
<br>
For this problem we refer to the following coordinate system (note different from problems 2 and 3). How you define your axis is arbitrary, but once you've defined your coordinate system, you must follow it.

a) Compute the total vertical stress assuming a lithostatic gradient of 23.8 MPa/km.

$$S_{v}=S_{33}=23.8\frac{MPa}{km}\frac{1km}{3280ft}7950ft = 57.67 MPa$$

b) Compute the effective vertical stress assuming hydrostatic pore pressure gradient

Assuming hydrostatic pore pressure gradient of 10 MPa/km or 0.433 psi/ft, the formation pressure would be 

$$P_p=10\frac{MPa}{km}\frac{1km}{3280ft}7950ft = 24.23 MPa$$

So effective stress at 7950 ft TVD would be

$$\sigma_{v}=\sigma_{33}=S_{v}-P_p = 57.67 MPa - 24.23 MPa =33.44 MPa$$

c) Compute horizontal effective stresses assuming linear isotropic elasticity, $v=0.22$ and that horizontal strains are nearly zero.

If $\epsilon_{H}=\epsilon_{11} = \epsilon_{22} = 0$, then $\sigma_{11} = \sigma_{22}$

From linear isotropic elasticity

$$\sigma_{11}=\sigma_{22}=\frac{v}{(1-v)}\sigma_{33}$$
<br>
$$\sigma_{H}=\frac{v}{(1-v)}\sigma_{V}=\frac{0.22}{(1-0.22)}33.44 = 9.43 MPa$$

d) Write out the tensor of effective stresses.

\begin{equation*}
\underline{\underline{\sigma}}=
\begin{vmatrix}
\sigma_{11} & \sigma_{12} & \sigma_{13} \\
\sigma_{21} & \sigma_{22} & \sigma_{23} \\
\sigma_{31} & \sigma_{32} & \sigma_{33} \\
\end{vmatrix}=
\begin{vmatrix}
\sigma_{H}  & 0 & 0 \\
0 & \sigma_{H} & 0 \\
0 & 0 & \sigma_{V} \\
\end{vmatrix}=
\begin{vmatrix}
9.43 & 0 & 0 \\
0 & 9.43 & 0 \\
0 & 0 & 33.44 \\
\end{vmatrix} MPa
\end{equation*}

e) Compute total horizontal stress.

\begin{equation*}
\underline{\underline{S}}=\underline{\underline{\sigma}}+\underline{\underline{P_p}}=
\begin{vmatrix}
9.43 & 0 & 0 \\
0 & 9.43 & 0 \\
0 & 0 & 33.44 \\
\end{vmatrix}+
\begin{vmatrix}
24.23 & 0 & 0 \\
0 & 24.23 & 0 \\
0 & 0 & 24.23 \\
\end{vmatrix}=
\begin{vmatrix}
33.66 & 0 & 0 \\
0 & 33.66 & 0 \\
0 & 0 & 57.67 \\
\end{vmatrix} MPa
\end{equation*}

f) Compute the ratio between effective horizontal stress and effective vertical stress

$$\frac{\sigma_{11} or\, \sigma_{22}}{\sigma_{33}}=\frac{\sigma_{H}}{\sigma_{V}}=\frac{9.43}{33.44}=0.28$$

This is the stress felt at faults

g) Compute the ratio between total horizontal stress and total vertical stress

$$\frac{S_{11} or\, S_{22}}{S_{33}}=\frac{S_{H}}{S_{V}}=\frac{33.66}{57.67}=0.60$$

h) Compute effective and total stresses assuming there is overpressure $\lambda_p=0.7$, tectonic strains $\epsilon_hmin=0$ and $\epsilon_hmax=0.0002$ and the shale Young’s modulus is $E=5 MMpsi$.

$$\lambda_p = \frac{P_p}{S_V}$$
<br>
$$P_p = \lambda_p \times S_V = 0.7 \times 57.67 MPa = 40.4 MPa$$

So effective vertical stress is

$$\sigma_V = S_V - P_p = 17.3 MPa$$

The maximum effective horizontal stress is given by

$$\sigma_{Hmax}=\frac{E}{1-v^2}\epsilon_{Hmax}+\frac{vE}{1-v^2}\epsilon_{Hmin}+\frac{v}{1-v}\sigma_{V}$$

The minimum horizontal stresses is given by

$$\sigma_{Hmin}=\frac{vE}{1-v^2}\epsilon_{Hmax}+\frac{E}{1-v^2}\epsilon_{Hmin}+\frac{v}{1-v}\sigma_{V}$$

With $E=5MM psi$ and $v=0.22$, the maximum effective horizontal stress is calculated to be

$$\sigma_{Hmax}=\frac{5000000 psi}{1-0.22^2} \times 0.0002+\frac{0.22 \times 5000000 psi}{1-0.22^2} \times 0+\frac{0.22}{1-0.22} \times 17.3 MPa = 7.21 MPa + 0 MPa + 4.88 MPa = 12.1 MPa$$ 

The minimum effective horizontal stress is calculated to be

$$\sigma_{Hmin}=\frac{0.22 \times 5000000 psi}{1-0.22^2} \times 0.0002+\frac{5000000 psi}{1-0.22^2} \times 0+\frac{0.22}{1- 0.22} \times 17.3 MPa = 1.58 MPa + 0 MPa + 4.88 MPa = 6.46 MPa $$

Note in above calculations, E must be in psi. 

In summary, effective stresses are

$$\sigma_{Hmax}=12.1MPa$$
<br>
$$\sigma_{Hmin}=6.46MPa$$
<br>
$$\sigma_V=17.3MPa$$

Total stresses are

$$S_{Hmax}=\sigma_{Hmax}+P_p=12.1MPa+40.4MPa=52.5MPa$$
<br>
$$S_{Hmin}=\sigma_{Hmin}+P_p=6.46MPa+40.4MPa=47.1MPa$$
<br>
$$S_V=\sigma_V+P_p=17.3 MPa+40.4MPa=57.5MPa$$
